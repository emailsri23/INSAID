```
df1 = pd.read_csv('https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/FastFoodRestaurants.csv')
df2 = df1.drop(columns='country')
df2
```

```
df2['name'].value_counts()
```

```
df2['name'].value_counts().plot.bar()
plt.xlabel('Shop Name')
plt.ylabel('No of Outlets')
plt.title('No. of outlets of each shop')
```

```
#df2['name'].value_counts().plot.hist()
df2['name'].value_counts().plot.bar(figsize=(20,7), yticks=np.arange(0,2000,100))
plt.xlabel('Shop Name')
plt.ylabel('No of Outlets')
plt.title('No. of outlets of each shop')
```
